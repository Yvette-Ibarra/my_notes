# $Seaborn$ $Data$ $Visualizations$

### Imports

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# import tips dataset
tips = sns.load_dataset('tips')
```

```python
# barplot with day and tip
sns.barplot(x='day',y='tip', data = tips);
```

```python
# barplot with day and total bill
sns.barplot(x='day',y='total_bill', data = tips);
```

### Set hue

```python
# barplot with hue and pallet
sns.barplot(x='day',y='total_bill', data = tips, hue='sex', palette= 'winter_r');
```

```python
# barplot with a hue
sns.barplot(x='day',y='total_bill', data = tips, hue = 'smoker');
```

```python
# Horizontal barplots switch x and y variables
sns.barplot(x='total_bill',y='day', data = tips, palette='spring');
```

```python
# order how the bars appear in the barplot
sns.barplot(x='day',y='tip', data = tips, palette='spring', order =['Sat','Fri','Sun','Thur']);
```

### Change estimator or statistic on barplot

```python
# change the estimator or statistic from mean to median
sns.barplot(x='day',y='total_bill', data = tips, estimator= np.median, palette='spring');
```

```python
# change the estimator or statistic from mean to median 
sns.barplot(x='smoker',y='tip', data = tips, estimator= np.median,hue='sex', palette='coolwarm');
```

### Set confidence intervals bars on barplots

```python
# set values for cofidence intervals using ci
sns.barplot(x='smoker',y='tip',data = tips, ci =99);
```

```python
# set values for cofidence intervals using ci
sns.barplot(x='smoker',y='tip',data = tips, ci =68);
```

```python
# set values for cofidence intervals using ci
sns.barplot(x='smoker',y='tip',data = tips, ci =30, palette='winter_r',estimator = np.median);
```

### Set capsize

```python
#Set capsize for bars
sns.barplot(x='day', y='total_bill', data = tips, palette = 'spring',capsize=0.3);
```

```python
#Set capsize for bars
sns.barplot(x='day', y='total_bill', data = tips, palette = 'spring',capsize=0.9);
```

```python
#Set capsize for bars
sns.barplot(x='day', y='total_bill', data = tips, palette = 'husl',capsize=0.1);
```

```python
#Set capsize for bars and hue
sns.barplot(x='day', y='total_bill', data = tips,hue='sex', palette = 'husl',capsize=0.1);
```

```python
# set autumn pallet and capsize
sns.barplot(x='size',y='tip', data=tips, palette='autumn',capsize=0.15);
```

```python
# bars have husl palette and capsize
sns.barplot(x='size',y='tip', data=tips, palette='husl',capsize=0.15);
```

```python
# bars are all one color
sns.barplot(x='size',y='tip', data=tips, color = 'green',capsize=0.15);
```
