## Line shape, line color, text

```python
x = np.linspace (-10,10,150)
cos = np.cos(x)
sin = np.sin(x)
```

```python
# types of lines
# : , -. -- , -
```

```python
# line style, color
plt.plot(x, cos, color = 'red', linewidth = 3.8, linestyle = ':')
```

```python
# line style and width
plt.plot(x, sin, linewidth = 3.8, linestyle = '-.')
```

```python
# line stype and color, two plots in one plane
plt.plot(x, sin, linewidth = 3.5,color = 'green', linestyle = '--')
plt.plot(x, cos ,linewidth = 3,color = 'red', linestyle = '--');
```

```python
# text with positive rotaion
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-4, -0.3, 'This text wil be displayed across our plot', rotation = 20, fontsize= 14);
```

```python
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-7, 0, 'This text wil be displayed across our plot', rotation = 35, fontsize= 14);
```

```python
# text with no rotaion
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-8, 0, 'This text wil be displayed across our plot', fontsize= 14);
```

```python
# text with negative rotaion
plt.plot(x, cos, color = 'green', linewidth = 3.2)

plt.text(-8, 0, 'This text wil be displayed across our plot',rotation = -20, fontsize= 14);
```

```python
# plot two plots at once verticaly
plt.subplot(211)
plt.plot(x, cos, color ='red', linewidth = 2.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation = 11, fontsize = 14)

plt.subplot(212)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation = -11, fontsize = 14)
```

```python
# change first number in subplot to go horizontal
plt.subplot(121)
plt.plot(x, cos, color ='red', linewidth = 2.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation = 11, fontsize = 14)

plt.subplot(122)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation =- 11, fontsize = 14)
```

```python
# define fiture for a better fit
plt.figure(figsize = (10,6))
plt.subplot(121)
plt.plot(x, cos, color ='red', linewidth = 2.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation = 11, fontsize = 14)

plt.subplot(122)
plt.plot(x, sin, color = 'green', linewidth = 3.8)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation =- 11, fontsize = 14)
```

## Labels, title, text and annotation

```python
x = np.linspace (-10,10,150)
cos = np.cos(x)
sin = np.sin(x)

plt.plot(x, cos, color = 'red', linewidth = 3.0);
```

```python
# add x and y labels
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis')
plt.ylabel('cosx');
```

```python
# add title
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis')
plt.ylabel('cosx')
plt.title('Graph of x vs cos(x)');
```

```python
# Add a grid
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis', fontsize = 15, color = 'green')
plt.ylabel('cosx', fontsize = 15, color = 'blue')
plt.title('Graph of x vs cos(x)')
plt.grid(True)
```

```python
# construct annotate
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis', fontsize = 15, color = 'green')
plt.ylabel('cosx', fontsize = 15, color = 'blue')
plt.title('Graph of x vs cos(x)')
plt.grid(True)

plt.annotate('this is one max value', xy = (0,1), xytext= (5, 0.5),
             arrowprops= dict(facecolor = 'black', shrink= 0.09));
```

```python
# change annotate position
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis', fontsize = 15, color = 'green')
plt.ylabel('cosx', fontsize = 15, color = 'blue')
plt.title('Graph of x vs cos(x)')
plt.grid(True)

plt.annotate('this is one max value', xy = (0,1), xytext= (7, 0.3),
             arrowprops= dict(facecolor = 'black', shrink= 0.14));
```

```python
# anotate on top of graph
plt.plot(x, cos, color = 'red', linewidth = 3.0)

plt.xlabel('X-axis', fontsize = 15, color = 'green')
plt.ylabel('cosx', fontsize = 15, color = 'blue')
plt.title('Graph of x vs cos(x)')
plt.grid(True)

plt.annotate('this is one max value', xy = (0,1), xytext= (7, 1.5),
             arrowprops= dict(facecolor = 'black', shrink= 0.14));
```

```python
x=np.linspace(-10, 10, 150)
cos = np.cos(x)
sin = np.sin(x)

plt.figure(figsize =(10,7))

plt.subplot(221)
plt.plot(x, cos, color= 'red', linewidth = 2.4, linestyle ='-.')
plt.title('cos(x)')
plt.grid(True)
plt.text(-5, -0.5, 'this will  run across our subplot', rotation = 10, fontsize = 15)

plt.subplot(222)

plt.plot(x, sin, color='green', linewidth = 3.4, linestyle = ':')
plt.title('sin(x)')
plt.grid(True)
plt.text(-5, -0.3, 'this will  run across our subplot', rotation = -10, fontsize = 15)

plt.subplot(223)
plt.plot(x,-cos, color = 'blue', linewidth = 3.5, linestyle='--')
plt.title('-cosx')
plt.grid(True)

plt.subplot(224)
plt.plot(x, -sin, color ='yellow', linewidth = 4.0, linestyle = '-')
plt.title('-sinx')
plt.grid(True);
```
